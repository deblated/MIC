## Варіант 7

\begin{equation}
\begin{cases}
    3.2x_1 + 8.8x_2 = 89 \\
    -14.2x_1 + 10.8x_2 = 125 \\
    20.3x_1 - 7.2x_x = 142 \\
\end{cases}
\end{equation} 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def eq1(x):
    return (89 - 3.2*x) / 8.8

def eq2(x):
    return (125 + 14.2*x) / 10.8

def eq3(x):
    return (142 - 20.3*x) / (-7.2)

x_values = np.linspace(-20, 20, 400)

plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.plot(x_values, eq1(x_values), color='red', linestyle='dashed')
plt.title(r'$3.2x_1 + 8.8x_2 = 89$')
plt.ylim(-20, 20)  # однакові межі для y

plt.subplot(132)
plt.plot(x_values, eq2(x_values), color='green', linestyle='dotted')
plt.title(r'$-14.2x_1 + 10.8x_2 = 125$')
plt.ylim(-20, 20)  # однакові межі для y

plt.subplot(133)
plt.plot(x_values, eq3(x_values), color='blue', linestyle='dashdot')
plt.title(r'$20.3x_1 - 7.2x_2 = 142$')
plt.ylim(-20, 20)  # однакові межі для y

plt.xlim(-20, 20)  # однакові межі для x

plt.tight_layout()
plt.savefig("myplot1.svg")
plt.savefig("myplot1.png")
plt.savefig("myplot1.jpg")
plt.show()
```

```python
x_values = np.linspace(-10, 25, 400)

plt.figure(figsize=(8, 16), dpi=100)
plt.grid(color='grey', linestyle='--')

plt.plot(x_values, eq1(x_values), color='red', linestyle='dashed', label='f[0]')
plt.text(0, eq1(0)-1.5, r'$3.2x_1 + 8.8x_2 = 89$', rotation=-27)

plt.plot(x_values, eq2(x_values), color='green', linestyle='dotted', label='f[1]')
plt.text(5, eq2(5)+2, r'$-14.2x_1 + 10.8x_2 = 125$', rotation=62)

plt.plot(x_values, eq3(x_values), color='blue', linestyle='dashdot', label='f[2]')
plt.text(16, eq3(15), r'$20.3x_1 - 7.2x_2 = 142$', rotation=75)

plt.ylim(0, 50)
plt.xlim(-10, 25)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')

plt.legend()

plt.title('Графіки системи рівнянь')


A = np.array([[3.2, 8.8], [-14.2, 10.8]])
B = np.array([89, 125])
x_intersection1, y_intersection1 = np.linalg.solve(A, B)
C = np.array([[-14.2, 10.8], [20.3, -7.2]])
D = np.array([125, 142])
x_intersection2, y_intersection2 = np.linalg.solve(C, D)
E = np.array([[3.2, 8.8], [20.3, -7.2]])
F = np.array([89, 142])
x_intersection3, y_intersection3 = np.linalg.solve(E, F)

plt.plot(x_intersection1, y_intersection1, 'mo', label='Intersection 1',markersize=10)
plt.plot(x_intersection2, y_intersection2, 'co', label='Intersection 2',markersize=10)
plt.plot(x_intersection3, y_intersection3, 'ro', label='Intersection 3',markersize=10)

x_values = [x_intersection1,x_intersection2,x_intersection3, x_intersection1]
y_values = [y_intersection1,y_intersection2,y_intersection3, y_intersection1]

plt.annotate("", xy=(x_intersection1, y_intersection1), xytext=(-5, 20), arrowprops=dict(arrowstyle="->"))
plt.annotate("Intersection between f[0] and f[1]", xy=(-9, 20.5))

plt.annotate("", xy=(x_intersection2, y_intersection2), xytext=(0, 35), arrowprops=dict(arrowstyle="->"))
plt.annotate("Intersection between f[1] and f[2]", xy=(-5, 34))

plt.annotate("", xy=(x_intersection3, y_intersection3), xytext=(20, 10), arrowprops=dict(arrowstyle="->"))
plt.annotate("Intersection between f[0] and f[2]", xy=(11.5, 10.5))

plt.fill_between(x_values, y_values, color='blue', alpha=0.3)

plt.savefig("myplot2.svg")
plt.savefig("myplot2.png")
plt.savefig("myplot2.jpg")
plt.show()

```
